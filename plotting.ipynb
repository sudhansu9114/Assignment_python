```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(11)
x
```

```python
y=x*2
y
```

```python
plt.plot(x,y)
plt.title("Line Plot of y = 2x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle='dashdot', linewidth=0.5, marker='o', markersize=5)
plt.title("Line Plot of y = 2x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
# Multiple lines in one graph

x=np.arange(0, 11)
y1=x*2
y2=x*3
plt.plot(x, y1, label='y = 2x', color='blue',linewidth=2)
plt.plot(x, y2, label='y = 3x', color='green', linewidth=5)
plt.title("Multiple Line Plot")
plt.xlabel("x values") 
plt.ylabel("y values")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Adding sub-plots

x=np.arange(0, 11)
y1=x*2
y2=x*3
plt.subplot(1, 2, 1) #plt.subplot(rows, columns, panel number)
plt.plot(x, y1, label='y = 2x', color='blue', linewidth=2)
plt.xlabel("x values")
plt.ylabel("y values")
plt.subplot(1, 2, 2)
plt.plot(x, y2, label='y = 3x', color='green', linewidth=5)

plt.suptitle("Subplots Example")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
# Bar plots

students={'John': 85, 'Alice': 90, 'Bob': 78, 'Diana': 92}
names=list(students.keys())
scores=list(students.values())
plt.bar(names, scores, color='#696969')
plt.title("Bar Plot of Student Scores")
plt.xlabel("Students") 
plt.ylabel("Scores")
plt.yticks(np.arange(0, 101, 10))
plt.xticks(rotation=45)
# plt.grid(axis='y')
plt.show()
```

```python
# Horizontal bar plot
plt.barh(names, scores, color='#696969')
plt.title("Horizontal Bar Plot of Student Scores")
plt.xlabel("Scores")
plt.ylabel("Students")
plt.xticks(np.arange(0, 101, 10))
plt.yticks(rotation=45)
plt.grid(axis='x')
plt.show()
```

```python
# Creating a scatter plot
x=np.arange(0, 20)
y=np.random.randint(0, 100, size=20)
plt.scatter(x, y, color='purple', marker='*', s=100)
plt.title("Scatter Plot Example")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.show()
```

```python
# Multiple scatter plots
x1=np.arange(0, 21)
y1=np.random.randint(0, 100, size=21)
y2=np.random.randint(0, 100, size=21)
plt.scatter(x1, y1, color='blue', label='Dataset 1', marker='o', s=50)
plt.scatter(x1, y2, color='red', label='Dataset 2', marker='x', s=50)
plt.title("Multiple Scatter Plots")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend(['Dataset 1', 'Dataset 2'])
plt.grid(True)
plt.show()
```

```python
# Subplots with scatter plots
x1=np.arange(0, 21)
y1=np.random.randint(0, 100, size=21)
y2=np.random.randint(0, 100, size=21)
plt.subplot(1, 2, 1)
plt.scatter(x1, y1, color='blue', label='Dataset 1', marker='o', s=50)
plt.title("Scatter Plot 1")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.subplot(1, 2, 2)
plt.scatter(x1, y2, color='red', label='Dataset 2', marker='x', s=50)
plt.title("Scatter Plot 2")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.suptitle("Subplots with Scatter Plots")
plt.show()
```

```python
# Histogram

data = np.random.randn(10000)  # Generate random data
plt.hist(data, bins=50, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
# Box plot
one=np.random.randint(1, 100, size=100)
two=np.random.randint(1, 100, size=100)
three=np.random.randint(1, 100, size=100)
data = [one, two, three]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.xlabel("Data Sets")
plt.ylabel("Values")
plt.xticks([1, 2, 3], ['Set 1', 'Set 2', 'Set 3'])
plt.grid(axis='y')
plt.show()
```

```python
# Violin plot
one=np.random.randint(1, 100, size=100)
two=np.random.randint(1, 100, size=100)
three=np.random.randint(1, 100, size=100)
plt.violinplot([one, two, three], showmeans=True)
plt.title("Violin Plot Example")
plt.xlabel("Data Sets")
plt.ylabel("Values")
plt.xticks([1, 2, 3], ['Set 1', 'Set 2', 'Set 3'])
plt.grid(axis='y')
plt.show()
```

```python
# Pie plot

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
quantities = [30, 20, 25, 15]
plt.pie(quantities, labels=fruits,autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Pie Chart of Fruit Quantities")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.pie([1],colors='white', radius=7)  # Draw a white circle in the center to create a donut chart effect
plt.show()
```
